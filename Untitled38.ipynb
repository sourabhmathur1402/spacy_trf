```python
class TrainDataGenerator(object):
    def __init__(self, text):
        self.text = text
        self.entities = []

    def add_entity(self, searchTerm = '', entity_name=''):
        try:
            response = re.search(searchTerm, self.text)
            data_entity = (response.start(), response.end(), entity_name)
            self.entities.append(data_entity)
        except Exception as e:pass

    def complete_entity(self):

        entity_tem = {"entities":self.entities}
        data = (self.text, entity_tem)

        entites = entity_tem.get("entities")

        # check if entity first index is overlapping with another one
        for i in range(0, len(entites)):
            for j in range(i+1, len(entites)):


                StartIndex1 = entites[i][0]
                endIndex1 = entites[i][1]

                StartIndex2 = entites[j][0]
                endIndex2 = entites[j][1]

                if StartIndex1 in range(StartIndex2, endIndex2):
                    return False
                if endIndex2 in range(StartIndex2, endIndex2):
                    return False

        return data
```

```python
import spacy
from spacy.training import Example
from spacy.util import minibatch, compounding
import random

# Load pre-trained transformer-based English model
nlp = spacy.load("en_core_web_trf")

class Model(object):
    def __init__(self, modelName="testmodel"):
        self.nlp = spacy.blank("en")  # Use a blank model
        self.modelName = modelName

    def train(self, TRAIN_DATA, output_dir=None, n_iter=80):
        # Check if 'ner' component already exists in the pipeline
        if 'ner' not in self.nlp.pipe_names:
            # Create a new NER component
            ner = self.nlp.add_pipe("ner")
        else:
            # Get the existing NER component
            ner = self.nlp.get_pipe("ner")

        # Add labels
        for _, annotations in TRAIN_DATA:
            for ent in annotations.get("entities"):
                ner.add_label(ent[2])

        # Get names of other pipes to disable them during training
        pipe_exceptions = ["ner", "trf_wordpiecer", "trf_tok2vec"]
        other_pipes = [pipe for pipe in self.nlp.pipe_names if pipe not in pipe_exceptions]

        with self.nlp.disable_pipes(*other_pipes):  # Only train NER
            optimizer = self.nlp.begin_training()

            for itn in range(n_iter):
                print("Iteration:", itn)
                random.shuffle(TRAIN_DATA)
                losses = {}

                # Batch up the examples using spaCy's minibatch
                batches = minibatch(TRAIN_DATA, size=compounding(4.0, 32.0, 1.001))

                for batch in batches:
                    examples = []
                    for text, annotations in batch:
                        examples.append(Example.from_dict(self.nlp.make_doc(text), annotations))
                    self.nlp.update(
                        examples,           # Batch of Example objects
                        drop=0.5,           # Dropout - make it harder to memorize data
                        losses=losses,
                    )

                print("Losses:", losses)

        if output_dir is not None:
            self.nlp.to_disk(output_dir)
            print(f"Model has been trained and saved in {output_dir}")
        else:
            print("Model has been trained.")
```

```python
def find_overlapping_entities(train_data):
    overlapping_examples = []

    for text, annotations in train_data:
        entities = annotations["entities"]
        for i, (start1, end1, label1) in enumerate(entities):
            for j, (start2, end2, label2) in enumerate(entities):
                if i != j and (start1 <= start2 < end1 or start1 < end2 <= end1):
                    overlapping_examples.append((text, entities))
                    break

    return overlapping_examples
```

```python
def remove_overlapping_examples(train_data):
    cleaned_train_data = []

    for text, annotations in train_data:
        entities = annotations["entities"]
        has_overlap = False
        for i, (start1, end1, label1) in enumerate(entities):
            for j, (start2, end2, label2) in enumerate(entities):
                if i != j and (start1 <= start2 < end1 or start1 < end2 <= end1):
                    has_overlap = True
                    break
            if has_overlap:
                break
        if not has_overlap:
            cleaned_train_data.append((text, annotations))

    return cleaned_train_data

# Find overlapping examples
overlapping_examples = find_overlapping_entities(TRAIN_DATA)

# Remove overlapping examples from TRAIN_DATA
cleaned_TRAIN_DATA = remove_overlapping_examples(TRAIN_DATA)

# Print the cleaned training data
for text, annotations in cleaned_TRAIN_DATA:
    print("Text:", text)
    print("Entities:", annotations)
    print()
```
